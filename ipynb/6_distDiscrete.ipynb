```python
p = 0.2
# The input has to be
bernoulliDist = stats.bernoulli(p)
```

```python
# Calculate the pmf
p_tails = bernoulliDist.pmf(0) # the sets the first variable the larger
p_heads = bernoulliDist.pmf(1)
```

## Binomial Distribution

```python
bd1 = stats.binom(20, 0.5)
bd2 = stats.binom(20, 0.7)
bd3 = stats.binom(40, 0.5)
k = arange(40)
plot(k, bd1.pmf(k), 'o-b')
plot(k, bd2.pmf(k), 'd-r')
plot(k, bd3.pmf(k), 's-g')
title('Binomial distribition')
legend(['p=0.5 and n=20', 'p=0.7 and n=20', 'p=0.5 and n=40'])
xlabel('X')
ylabel('P(X)')
```

```python
# To claulate the chances of heads coming up over  n number of times we need the probability and the number of samples
# This specifically is the chance of a  success of the given variable
(p, num) = (0.5, 2)
binomDist = stats.binom(num, p)
```

```python
three_results = binomDist.pmf(np.arange(3)) # to get this to run do n+1 for the number of rols
# alright so what this gives you is from left to right the chances that heads comes up
```

```python
for i,v in enumerate(three_results):
    print(f'If rolled {i+1} times, the chances of heads is {v* 100}%')
```

The big difference is that while binomial distributions gives you the chances of a "this happened on this turn as a success" vs  Poisson. Poisson test would be such as " how many pennies will i encounter on the way home"

## Poisson Distribution

### Different Views of the Poisson Distrubution

```python
k = arange(30)
plot(k, pd.cdf(k))
title('Poisson distribition - CDF')
xlabel('X')
ylabel('P(X)')
```

```python
y = linspace(0,1,100)
plot(y, pd.ppf(y))
title('Poisson distribition - PPF')
xlabel('X')
ylabel('P(X)')
```
